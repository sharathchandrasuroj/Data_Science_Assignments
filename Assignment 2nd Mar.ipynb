### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.


Matplotlib:

Matplotlib is a 2D plotting library for Python that produces high-quality charts, figures, and visualizations. It provides an object-oriented API for embedding plots into applications that use general-purpose GUI toolkits such as Tkinter, PyQt, and wxPython, as well as for generating static images for use in web applications or scientific publications.

Why is it used:

Matplotlib is used for visualizing data and creating a wide variety of plots and charts. Some of its key features include:


Versatility: Matplotlib supports a wide variety of plot types and customization options, making it suitable for various data visualization needs.


Publication-Quality Plots: It allows the creation of publication-quality plots with fine control over the appearance of plots, labels, and annotations.


Integration: It seamlessly integrates with Jupyter Notebooks and other interactive environments.


Extensibility: Users can extend its functionality and create complex visualizations by combining Matplotlib with other libraries.


Community Support: Being widely used, Matplotlib has a large and active community, which means plenty of resources and examples are available for users.

```python
#Line Plot:

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```python
#Scatter Plot:

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
#Bar Chart:

import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C']
values = [10, 20, 15]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

```python
#Histogram:

import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# Pie Chart:

import matplotlib.pyplot as plt

labels = ['Category A', 'Category B', 'Category C']
sizes = [30, 45, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart')
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
#Bar plot is used to show plot categorical values or where we have some name or string on xaxis and respective value of yaxis

plt.bar(company,profit)
plt.xlabel("This are companies")
plt.ylabel("This are profits")
plt.title("This is my graph")
```
